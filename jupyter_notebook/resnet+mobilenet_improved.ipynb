```python
# =============================================================================
# CELL 1: IMPORTS AND SETUP
# =============================================================================
import pandas as pd
import numpy as np
import tensorflow as tf
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_absolute_error
import joblib
import os
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from IPython.display import display

# Suppress TensorFlow/Keras warnings for a cleaner output
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
print("✅ Libraries imported successfully.")


# =============================================================================
# CELL 2: CONFIGURATION
# =============================================================================
# !!! ATTENTION: PLEASE VERIFY THESE TWO PATHS ARE CORRECT !!!

# 1. This path should point to your CLEAN, 471-image CSV file.
CSV_FILE_PATH = "C:/Users/DELL/Downloads/banana_days_471_synced.csv"

# 2. This path must point to the REAL location of your 'images' folder.
IMAGES_DIR = 'C:/banana_images_jpg'

# --- Other settings ---
MODELS_DIR = 'models_ensemble_hybrid'
RESNET_FE_NAME = os.path.join(MODELS_DIR, 'cnn_resnet50_feature_extractor.keras')
MOBILENET_FE_NAME = os.path.join(MODELS_DIR, 'cnn_mobilenetv2_feature_extractor.keras')
LGBM_RESNET_BASE = os.path.join(MODELS_DIR, 'lgbm_resnet_fold_{}.pkl')
LGBM_MOBILENET_BASE = os.path.join(MODELS_DIR, 'lgbm_mobilenet_fold_{}.pkl')
IMG_WIDTH, IMG_HEIGHT = 224, 224
BATCH_SIZE = 32
N_SPLITS = 5 # Using 5-Fold Cross-Validation for robustness

os.makedirs(MODELS_DIR, exist_ok=True)
print("✅ Configuration set. Please ensure the paths above are correct before proceeding.")


# =============================================================================
# CELL 3: DATA LOADING AND VERIFICATION
# =============================================================================
print("\n--- Verifying Data Integrity ---")

if not os.path.exists(CSV_FILE_PATH):
    print(f"❌ FATAL ERROR: The CSV file was not found at '{CSV_FILE_PATH}'")
    sys.exit(1)

try:
    df = pd.read_csv(CSV_FILE_PATH, encoding='latin1', sep=',')
    if 'image_filename' not in df.columns or 'days_left' not in df.columns:
        print("❌ FATAL ERROR: CSV must contain 'image_filename' and 'days_left' columns.")
        sys.exit(1)
except pd.errors.ParserError as e:
    print(f"❌ FATAL CSV PARSING ERROR: {e}")
    sys.exit(1)

df['image_filename_full'] = df['image_filename'].apply(lambda x: os.path.join(IMAGES_DIR, x))
missing_files = [f for f in df['image_filename_full'] if not os.path.exists(f)]

if missing_files:
    print(f"\n❌ FATAL ERROR: {len(missing_files)} image files from your CSV were not found in the specified IMAGES_DIR!")
    print(f"Example missing file: {missing_files[0]}")
    sys.exit(1)

print(f"✅ SUCCESS: Loaded and verified {len(df)} records.")


# =============================================================================
# CELL 4: DATA ANALYSIS
# =============================================================================
print("\n--- Analyzing Data Distribution ---")
plt.figure(figsize=(12, 7))
sns.histplot(df['days_left'], bins=df['days_left'].nunique(), kde=False, color='orange')
plt.title('CRITICAL ANALYSIS: Data Distribution of Your Dataset', fontsize=16)
plt.xlabel('Days Left (Categories)', fontsize=12)
plt.ylabel('Number of Images (Frequency)', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.xticks(sorted(df['days_left'].unique()))
plt.show()



# =============================================================================
# CELL 5: BUILD AND EXTRACT FEATURES FROM BOTH CNNS
# =============================================================================
def build_feature_extractor(base_model_class, input_shape):
    """A helper function to build a CNN feature extractor."""
    base_model = base_model_class(input_shape=input_shape, include_top=False, weights='imagenet')
    base_model.trainable = False
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    return tf.keras.Model(inputs=base_model.input, outputs=x)

print("\n--- Building MobileNetV2 & ResNet50 Feature Extractors ---")
fe_mobilenet = build_feature_extractor(tf.keras.applications.MobileNetV2, (IMG_WIDTH, IMG_HEIGHT, 3))
fe_resnet = build_feature_extractor(tf.keras.applications.ResNet50, (IMG_WIDTH, IMG_HEIGHT, 3))
print("✅ CNN Feature Extractors built.")

def extract_features(dataframe, model, preprocess_func):
    """A helper function to extract features for a given CNN."""
    filenames = dataframe['image_filename_full'].tolist()
    dataset = tf.data.Dataset.from_tensor_slices(filenames)
    def parse_image(filename):
        image = tf.io.read_file(filename)
        image = tf.io.decode_jpeg(image, channels=3)
        image = tf.image.resize(image, [IMG_WIDTH, IMG_HEIGHT])
        image = preprocess_func(image)
        return image
    dataset = dataset.map(parse_image).batch(BATCH_SIZE)
    return model.predict(dataset, verbose=1)

print("\n--- Extracting Features from All Images (this may take a few minutes) ---")
print("\nExtracting with MobileNetV2...")
features_mobilenet = extract_features(df, fe_mobilenet, tf.keras.applications.mobilenet_v2.preprocess_input)
print("\nExtracting with ResNet50...")
features_resnet = extract_features(df, fe_resnet, tf.keras.applications.resnet50.preprocess_input)
all_labels = df['days_left'].values
print(f"✅ Feature extraction complete.")


# =============================================================================
# CELL 6: K-FOLD CROSS-VALIDATION TRAINING FOR DUAL ENSEMBLE
# =============================================================================
print(f"\n--- Starting {N_SPLITS}-Fold Cross-Validation Training for Dual Ensemble ---")
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
models_resnet, models_mobilenet = [], []
scores_resnet, scores_mobilenet = [], []

# --- THIS IS THE FIX ---
# We use `features_resnet` to get the correct number of samples for the split.
# The indices (train_idx, val_idx) will be the same for all feature sets.
for fold, (train_idx, val_idx) in enumerate(skf.split(features_resnet, all_labels)):
    print(f"\n===== FOLD {fold+1}/{N_SPLITS} =====")
    
    # --- Train ResNet Branch ---
    X_train_res, X_val_res = features_resnet[train_idx], features_resnet[val_idx]
    y_train, y_val = all_labels[train_idx], all_labels[val_idx]
    lgbm_res = lgb.LGBMRegressor(objective='mae', metric='mae', n_estimators=1000, learning_rate=0.05, random_state=42, n_jobs=-1)
    lgbm_res.fit(X_train_res, y_train, eval_set=[(X_val_res, y_val)], callbacks=[lgb.early_stopping(100, verbose=False)])
    preds_res = lgbm_res.predict(X_val_res)
    mae_res = mean_absolute_error(y_val, preds_res)
    scores_resnet.append(mae_res)
    models_resnet.append(lgbm_res)
    print(f"ResNet Branch Fold {fold+1} MAE: {mae_res:.4f}")
    
    # --- Train MobileNetV2 Branch ---
    X_train_mob, X_val_mob = features_mobilenet[train_idx], features_mobilenet[val_idx]
    lgbm_mob = lgb.LGBMRegressor(objective='mae', metric='mae', n_estimators=1000, learning_rate=0.05, random_state=42, n_jobs=-1)
    lgbm_mob.fit(X_train_mob, y_train, eval_set=[(X_val_mob, y_val)], callbacks=[lgb.early_stopping(100, verbose=False)])
    preds_mob = lgbm_mob.predict(X_val_mob)
    mae_mob = mean_absolute_error(y_val, preds_mob)
    scores_mobilenet.append(mae_mob)
    models_mobilenet.append(lgbm_mob)
    print(f"MobileNetV2 Branch Fold {fold+1} MAE: {mae_mob:.4f}")
    
# =============================================================================
# CELL 7: FINAL EVALUATION AND MODEL SAVING
# =============================================================================
print("\n--- Cross-Validation Training Complete ---")
mean_mae_res = np.mean(scores_resnet)
mean_mae_mob = np.mean(scores_mobilenet)
# Calculate the overall average MAE for the combined ensemble
overall_avg_mae = np.mean(scores_resnet + scores_mobilenet)

print(f"\n=======================================================")
print(f"  Average ResNet50 Validation MAE: {mean_mae_res:.4f}")
print(f"  Average MobileNetV2 Validation MAE: {mean_mae_mob:.4f}")
print(f"  --------------------------------------------------")
print(f"  Expected Final Ensemble MAE is approx: {overall_avg_mae:.4f}")
print(f"=======================================================")

# Save all 12 model components
fe_resnet.save(RESNET_FE_NAME)
fe_mobilenet.save(MOBILENET_FE_NAME)
print(f"✅ CNN Feature Extractors saved.")
for i in range(N_SPLITS):
    joblib.dump(models_resnet[i], LGBM_RESNET_BASE.format(i+1))
    joblib.dump(models_mobilenet[i], LGBM_MOBILENET_BASE.format(i+1))
print(f"✅ All 10 LightGBM models saved.")


# =============================================================================
# CELL 8: ENSEMBLE PREDICTION PIPELINE
# =============================================================================
print("\n--- Final Ensemble Prediction Pipeline Ready ---")
# Load all models for prediction
predictor_resnet_cnn = tf.keras.models.load_model(RESNET_FE_NAME)
predictor_mobilenet_cnn = tf.keras.models.load_model(MOBILENET_FE_NAME)
predictor_lgbm_res, predictor_lgbm_mob = [], []
for i in range(N_SPLITS):
    predictor_lgbm_res.append(joblib.load(LGBM_RESNET_BASE.format(i+1)))
    predictor_lgbm_mob.append(joblib.load(LGBM_MOBILENET_BASE.format(i+1)))

def predict_ensemble(image_path):
    try:
        img = Image.open(image_path).resize((IMG_WIDTH, IMG_HEIGHT))
        img_array = np.array(img)[:, :, :3]
        img_batch = np.expand_dims(img_array, axis=0)
        
        # Get features from both CNNs
        features_res = predictor_resnet_cnn.predict(tf.keras.applications.resnet50.preprocess_input(img_batch.copy()), verbose=0)
        features_mob = predictor_mobilenet_cnn.predict(tf.keras.applications.mobilenet_v2.preprocess_input(img_batch.copy()), verbose=0)
        
        # Get predictions from all 10 LGBM models
        preds_res = [model.predict(features_res)[0] for model in predictor_lgbm_res]
        preds_mob = [model.predict(features_mob)[0] for model in predictor_lgbm_mob]
        
        # Average all predictions for the final result
        final_prediction = np.mean(preds_res + preds_mob)
        return final_prediction
    except FileNotFoundError: return "Error: File not found."
    except Exception as e: return f"An error occurred: {e}"

print(f"✅ All 12 models loaded for ensemble prediction.")
```
